```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python


def scrape_data_Name():
  b=[]
  names = soup.find_all('a',class_ = "course-card course-card__public published")
  for name in names: # iterate through the elements in the list 'names'
    h3_tags = name.find_all("h3")  # find h3 tags within each 'name' element
  #print(h3_tags)
  # Now you can do something with the h3_tags, e.g., extract their text
    for h3 in h3_tags:
      try :
        b.append(h3.text.strip())
      except:
        b.append(np.nan)
  return b

def scrape_data_Link():
  a = [] # course link save hogi yha
  for link in soup.find_all('a', class_ = "course-card course-card__public published"):
    href_value = link.get('href')
    if href_value:  # Check if href_value is not None
      try :
        a.append(str(href_value))
      except:
        a.append(np.nan)
  return a

def scrape_data_Description():
  d = []
  #soup.find_all('div',class_="fr-view")[0].text.strip()
  try :
    d.append(soup.find('div', class_="fr-view").text.strip() if soup.find('div', class_="fr-view") else np.nan)
  except:
    d.append(np.nan)

  return d

def scrape_data_Curriculum():
  c = []
  x = []
  for i in soup.find_all('span',class_="course-curriculum__chapter-lesson"):
    x.append(i.text.strip())
  c.append(x)
  if c == []:
    c = [["NAN"]]
  #print(c)
  return c


final = pd.DataFrame()

curr = []
name = []
desc = []
link = []
url1 = []
for j in range(1,10):
  url = 'https://courses.analyticsvidhya.com/collections/courses?page='+str(j)
  url1.append(url)


for url in url1:
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  response = requests.get(url, headers=headers)
  text = response.text
  soup=BeautifulSoup(text,'lxml')
  name = name + scrape_data_Name()
  link = link + scrape_data_Link()
  #print(soup)


base_link = "https://courses.analyticsvidhya.com/"

for i in range(len(link)):
  a = link[i]


  url = base_link + a
  #print(url)
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  response = requests.get(url, headers=headers)
  text = response.text
  soup=BeautifulSoup(text,'lxml')

  curr = curr + scrape_data_Curriculum()
  desc = desc + scrape_data_Description()




```
